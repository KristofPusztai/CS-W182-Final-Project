```python
import pickle
import matplotlib.pyplot as plt

timestep_array = pickle.load(open('/content/gdrive/MyDrive/CSW182 Final/50FruitBot/50timestep_array.pickle', 'rb'))
score_array = pickle.load(open('/content/gdrive/MyDrive/CSW182 Final/50FruitBot/50score_array.pickle', 'rb'))
reg_timestep_array = pickle.load(open('/content/gdrive/MyDrive/CSW182 Final/50FruitBot/reg50timestep_array.pickle', 'rb'))
reg_score_array = pickle.load(open('/content/gdrive/MyDrive/CSW182 Final/50FruitBot/reg50score_array.pickle', 'rb'))
dec_score_array = pickle.load(open('/content/gdrive/MyDrive/CSW182 Final/50FruitBot/dec50score_array.pickle', 'rb'))
dec_timestep_array = pickle.load(open('/content/gdrive/MyDrive/CSW182 Final/50FruitBot/dec50timestep_array.pickle', 'rb'))

if len(timestep_array) > len(reg_timestep_array):
  if len(dec_timestep_array) > len(reg_timestep_array):
    plt_timestep = reg_timestep_array
  else:
    plt_timestep = dec_timestep_array
elif len(dec_timestep_array) > len(timestep_array):
  plt_timestep = timestep_array
else:
   plt_timestep = dec_timestep_array
reg_score_array = reg_score_array[:len(plt_timestep)]
score_array = score_array[:len(plt_timestep)]
dec_score_array = dec_score_array[:len(plt_timestep)]

plt.plot(plt_timestep, score_array)
plt.plot(plt_timestep, reg_score_array)
plt.plot(plt_timestep, dec_score_array)
plt.title('Training Reward for 50 Level Environment')
plt.xlabel('Total Timesteps')
plt.ylabel('Mean reward over 100 different levels')
plt.legend(['PPO2', 'MIXREG', 'DECOUPLED'])
plt.show;
```
